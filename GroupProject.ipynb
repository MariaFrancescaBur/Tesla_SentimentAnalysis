```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Apply sentiment analysis
TeslaBIGV2['sentiment'] = TeslaBIGV2['full_text'].apply(lambda text: sia.polarity_scores(text)['compound'])

# Plot sentiment distribution
plt.figure(figsize=(8, 6))
sns.histplot(TeslaBIGV2['sentiment'], bins=30, kde=True, color='purple')
plt.title("Sentiment Distribution of Tweets")
plt.xlabel("Sentiment Score")
plt.ylabel("Frequency")
plt.show()
```

```python
def classify_sentiment(score):
    if score > 0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

TeslaBIGV2['sentiment_category'] = TeslaBIGV2['sentiment'].apply(classify_sentiment)
print(TeslaBIGV2['sentiment_category'].value_counts())
```

```python
sentiment_counts = TeslaBIGV2['sentiment_category'].value_counts()

plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'gray', 'red'])
plt.title("Sentiment Categories of Tweets")
plt.xlabel("Sentiment Category")
plt.ylabel("Number of Tweets")
plt.xticks(rotation=0)
plt.show()
```

```python
# filter for truck and trucks and Truck Trucks; Pickup Pickuptruck...
keywords = ['Cybertruck', 'cybertruck', 'Cybertrucks', 'cybertrucks']
TeslaBIGV2['related_to_cybertruck'] = TeslaBIGV2['full_text'].apply(
    lambda x: any(keyword in x for keyword in keywords)
)

cybertruck_data = TeslaBIGV2[TeslaBIGV2['related_to_cybertruck']]
general_tesla_data = TeslaBIGV2[~TeslaBIGV2['related_to_cybertruck']]

# Perform sentiment analysis separately for both
cybertruck_data['sentiment'] = cybertruck_data['full_text'].apply(
    lambda x: sia.polarity_scores(x)['compound']
)
general_tesla_data['sentiment'] = general_tesla_data['full_text'].apply(
    lambda x: sia.polarity_scores(x)['compound']
)

# Visualize sentiment comparison
plt.figure(figsize=(10, 5))
sns.histplot(cybertruck_data['sentiment'], color='blue', label='Cybertruck', kde=True, bins=30)
sns.histplot(general_tesla_data['sentiment'], color='orange', label='Other Tesla Topics', kde=True, bins=30)
plt.legend()
plt.title("Sentiment Distribution: Cybertruck vs General Tesla sentiment")
plt.xlabel("Sentiment Score")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define thresholds for sentiment categories
def classify_sentiment(score):
    if score > 0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Classify sentiment for Cybertruck-related and general Tesla tweets
cybertruck_data['sentiment_category'] = cybertruck_data['sentiment'].apply(classify_sentiment)
general_tesla_data['sentiment_category'] = general_tesla_data['sentiment'].apply(classify_sentiment)

# Calculate proportions for each sentiment category
cybertruck_sentiment_counts = cybertruck_data['sentiment_category'].value_counts(normalize=True)
general_tesla_sentiment_counts = general_tesla_data['sentiment_category'].value_counts(normalize=True)

# Plot pie chart for Cybertruck sentiment
plt.figure(figsize=(8, 6))
cybertruck_sentiment_counts.plot.pie(
    autopct='%1.1f%%', startangle=140, colors=['green', 'gray', 'red'], labels=['Positive', 'Neutral', 'Negative']
)
plt.title("Sentiment Distribution for Cybertruck Tweets")
plt.ylabel("")  # Hide the ylabel
plt.show()

# Plot pie chart for General Tesla sentiment
plt.figure(figsize=(8, 6))
general_tesla_sentiment_counts.plot.pie(
    autopct='%1.1f%%', startangle=140, colors=['green', 'gray', 'red'], labels=['Positive', 'Neutral', 'Negative']
)
plt.title("Sentiment Distribution for General Tesla Tweets")
plt.ylabel("")  # Hide the ylabel
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Filter Cybertruck tweets
cybertruck_tweets = TeslaBIGV2[TeslaBIGV2['full_text'].str.contains('Cybertruck', case=False, na=False)]

# Preprocess text
vectorizer = CountVectorizer(max_df=0.9, min_df=10, stop_words='english')
dtm = vectorizer.fit_transform(cybertruck_tweets['full_text'])

# Apply LDA
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(dtm)

# Print topics
for index, topic in enumerate(lda.components_):
    print(f"Topic #{index}:")
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
```

```python
# Top engaging tweets related to Cybertruck
cybertruck_tweets = TeslaBIGV2[TeslaBIGV2['full_text'].str.contains('Cybertruck', case=False, na=False)]
top_engaging = cybertruck_tweets.sort_values(by=['retweet_count', 'favorite_count'], ascending=False).head(10)

print(top_engaging[['full_text', 'retweet_count', 'favorite_count']])
```

```python
cybertruck_tweets['month'] = cybertruck_tweets['created_at'].dt.to_period('M')
sentiment_trend = cybertruck_tweets.groupby('month')['sentiment'].mean()

plt.figure(figsize=(10, 6))
sentiment_trend.plot(kind='line', marker='o', color='green')
plt.title("Sentiment Trend Over Time for Cybertrucks")
plt.xlabel("Month")
plt.ylabel("Average Sentiment")
plt.grid()
plt.show()
```

```python
positive_tweets = cybertruck_tweets[cybertruck_tweets['sentiment'] > 0.05]
negative_tweets = cybertruck_tweets[cybertruck_tweets['sentiment'] < -0.05]

# Most common hashtags for positive and negative tweets
positive_hashtags = [hashtag for hashtags in positive_tweets['hashtags_extracted'].dropna() for hashtag in hashtags]
negative_hashtags = [hashtag for hashtags in negative_tweets['hashtags_extracted'].dropna() for hashtag in hashtags]

print("Top Positive Hashtags:", Counter(positive_hashtags).most_common(10))
print("Top Negative Hashtags:", Counter(negative_hashtags).most_common(10))
```

```python
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 hashtags for positive and negative tweets
positive_hashtags_count = Counter(positive_hashtags).most_common(10)
negative_hashtags_count = Counter(negative_hashtags).most_common(10)

# Convert to DataFrame for visualization
positive_hashtags_df = pd.DataFrame(positive_hashtags_count, columns=["Hashtag", "Frequency"])
negative_hashtags_df = pd.DataFrame(negative_hashtags_count, columns=["Hashtag", "Frequency"])

# Plot for Positive Hashtags
plt.figure(figsize=(10, 6))
sns.barplot(data=positive_hashtags_df, x="Frequency", y="Hashtag", palette="Greens_d")
plt.title("Top 10 Positive Hashtags for Cybertruck Tweets")
plt.xlabel("Frequency")
plt.ylabel("Hashtag")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Plot for Negative Hashtags
plt.figure(figsize=(10, 6))
sns.barplot(data=negative_hashtags_df, x="Frequency", y="Hashtag", palette="Reds_d")
plt.title("Top 10 Negative Hashtags for Cybertruck Tweets")
plt.xlabel("Frequency")
plt.ylabel("Hashtag")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
